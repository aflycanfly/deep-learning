# tf.image进行增强

```python
cat = plt.imread('./cat.jpg')
```

```python
plt.imshow(cat)
```

```python
cat1 = tf.image.random_flip_left_right(cat)
plt.imshow(cat1)
```

```python
cat2 = tf.image.random_flip_up_down(cat)
plt.imshow(cat2)
```

```python
cat3 = tf.image.random_crop(cat,(200,200,3))
plt.imshow(cat3)
```

```python
cat4 = tf.image.random_brightness(cat,0.5)
plt.imshow(cat4)
```

```python
cat5 = tf.image.random_hue(cat,0.5)
plt.imshow(cat5)
```

# 使用imagedataGenerator进行增强

```python
from tensorflow.keras.datasets import mnist
```

```python
# 获取数据集
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# 将数据转换为4维的形式
x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)
```

```python
# 实例化
datagen = tf.keras.preprocessing.image.ImageDataGenerator(shear_range=10)
```

```python
for x,y in datagen.flow(x_train,y_train,batch_size=9):
    plt.figure(figsize=(8,8))
    for i in range(0,9):
        plt.subplot(330+1+i)
        plt.imshow(x[i].reshape(28,28),cmap='gray')
        plt.title(y[i])
    plt.show()
    break
```
